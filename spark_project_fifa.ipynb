```
dfs.createOrReplaceTempView("DEFENSE")

display(spark.sql("""
                    WITH BEST_TW AS (
                    SELECT team, position, max(tackles_won) AS MAX_TW
                    FROM DEFENSE 
                    GROUP BY team, position
                    )
                   
                   SELECT 
                       o.player,
                       s.team,
                       s.position,
                       s.MAX_TW
                   FROM BEST_TW s
                   LEFT JOIN DEFENSE o
                   ON o.team = s.team AND o.position = s.position AND o.tackles_won = s.MAX_TW
                   WHERE s.position <> "GK"
                   ORDER BY s.team, s.position, s.MAX_TW 
                   
                   """))
```

```
dfs.createOrReplaceTempView("DEFENSE")

display(spark.sql("""
                    WITH BEST_TW AS (
                    SELECT 
                        team,
                        position,
                        max(tackles_won) AS MAX_TW
                    FROM DEFENSE 
                    GROUP BY team, position
                    ),
                    RN_BEST_TW AS (
                    SELECT 
                        s.team,
                        s.position,
                        o.player,
                        s.MAX_TW,
                        ROW_NUMBER() OVER(PARTITION BY s.team ORDER BY s.MAX_TW DESC) AS R_NUMBER
                        
                    FROM BEST_TW s
                    JOIN DEFENSE o
                    ON o.team = s.team AND o.position = s.position AND o.tackles_won = s.MAX_TW
                    )
                   
                   SELECT 
                       team, player, position, max_tw
                   FROM RN_BEST_TW 
                   WHERE R_NUMBER IN (1,2,3)
                       AND team in ("Argentina", "Croatia", "Morocco", "France")
                   ORDER BY MAX_TW DESC
                   
                   """))
```

```
gca.createOrReplaceTempView("GCA")
stats.createOrReplaceTempView("STATS")

display(spark.sql("""
                    WITH BEST_SCA AS (
                    SELECT 
                        team,
                        position,
                        max(sca) AS MAX_SCA
                    FROM GCA 
                    GROUP BY team, position
                    ),
                    RN_BEST_SCA AS (
                    SELECT 
                        s.team,
                        s.position,
                        o.player,
                        s.MAX_SCA,
                        ROW_NUMBER() OVER(PARTITION BY s.team ORDER BY s.MAX_SCA DESC) AS R_NUMBER
                        
                    FROM BEST_SCA s
                    JOIN GCA o
                    ON o.team = s.team AND o.position = s.position AND o.sca = s.MAX_SCA
                    )
                   
                   SELECT 
                       RN.team, RN.player, RN.position, RN.max_sca, st.goals, st.assists, 
                       st.goals + st.assists AS ASSISTorGOAL, ROUND(1/((st.goals + st.assists)/RN.max_sca), 3) AS Num_SituatiationLeadAorG 
                   FROM RN_BEST_SCA RN
                   JOIN STATS ST
                   ON st.player = rn.player
                   WHERE R_NUMBER IN (1,2,3)
                       AND RN.team in ("Argentina", "Croatia", "Morocco", "France")
                   ORDER BY MAX_sca DESC
                   
                   """))

#Num_SituatiationLeadAorG pecentage of situation that the player needs to lead a goal
#Perisic, Mbappe e Messi are the ones who need few chances to score or to assist 
```
